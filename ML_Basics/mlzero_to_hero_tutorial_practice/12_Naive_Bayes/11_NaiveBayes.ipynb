# Naive Bayes Classifier

Naive Bayes classifiers are a collection of classification algorithms based on Bayes' Theorem. It is not a single algorithm but a family of algorithms where all of them share a common principle, i.e. every pair of features being classified is independent of each other.

For example, a fruit may be considered to be an apple if it is red, round, and about 3 inches in diameter. Even if these features depend on each other or upon the existence of the other features, all of these properties independently contribute to the probability that this fruit is an apple and that is why it is known as ‘Naive’.





## Applications:

Naïve Bayes is one of the most straightforward and fast classification algorithm. It is very well suited for large volume of data. It is successfully used in various applications such as :
1. Spam filtering
2. Text classification
3. Sentiment analysis
4. Recommender systems

It uses Bayes theorem of probability for prediction of unknown class.

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import preprocessing
```

```python
data = pd.read_csv('credit_data.csv')
data.head()
```

```python
data.shape
```

## Feature Extraction

```python
features = data[['income', 'age', 'loan']]
target = data.default

feature_train, feature_test, target_train, target_test = train_test_split(features, target)
```

## Training

```python
model = GaussianNB()
model = model.fit(feature_train, target_train)
prediction = model.predict(feature_test)
```

## Confusion Matrix and Accuracy Score

```python
print(confusion_matrix(target_test, prediction))
print(accuracy_score(target_test, prediction))
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(target_test, prediction), annot=True, cmap='Blues')
```
